image.png

* vector --> Matrix 로 변하면 axis가 변화 하는것이다. 


* axis 벡터의 기준이다. 

```python
test_array = np.arange(1,13).reshape(3,4)
test_array
```

```python
test_array.sum()
```

```python
test_array.sum(axis=1)
```

```python
test_array.sum(axis=0)
```

## axis = 0, 1, 2

```python
third_order_tensor = np.array([test_array,test_array,test_array])
third_order_tensor
```

```python
a0 = third_order_tensor.sum(axis=0)
print(a0,'\n',a0.shape)
```

```python
a1 = third_order_tensor.sum(axis=1)
print(a1,'\n',a1.shape)
```

```python
a2 = third_order_tensor.sum(axis=2)
print(a2,'\n',a2.shape)
```

```python
a1 = np.arange(1, 10)
print(a1)
b1 = np.random.randint(1, 10, size=9)
print(b1)
print(a1 + b1)
print(a1 - b1)
print(a1 * b1)
print(a1 / b1)
print(a1 // b1)
print(a1 ** b1)
print(a1 % b1)
```

## sum(): 합 계산

```python
a1 = np.arange(1, 10).reshape(3, 3)
print(a1)
b1 = np.random.randint(1, 10, size=(3, 3))
print(b1)
print(a1 + b1)
print(a1 - b1)
print(a1 * b1)
print(a1 / b1)
print(a1 // b1)
print(a1 ** b1)
print(a1 % b1)
```

```python
a2 = np.random.randint(1, 10, size=(3, 3))
print(a2)
print(a2.sum(), np.sum(a2))
print(a2.sum(axis=0), np.sum(a2, axis=0))
print(a2.sum(axis=1), np.sum(a2, axis=1))
```

```python
a1 = np.random.randint(1, 10, size=5)
print(a1)
print(np.exp(a1))

print(np.exp2(a1))
print(np.power(a1, 2))
```

```python
print(a1)
print(np.log(a1))
print(np.log2(a1))
print(np.log10(a1))
```

## cumsum(): 누적합 계산

```python
print(a2)
print( np.cumsum(a2))
print( np.cumsum(a2, axis=0))
print( np.cumsum(a2, axis=1))
```

## diff(): 차분 계산

```python
print(a2)
print( np.diff(a2))
print( np.diff(a2, axis=0))
print( np.diff(a2, axis=1))
```

## prod(): 곱 계산

```python
print(a2)
print( np.prod(a2))
print( np.prod(a2, axis=0))
print( np.prod(a2, axis=1))
```

## cumprod(): 누적곱 계산

```python
print(a2)
print( np.cumprod(a2))
print( np.cumprod(a2, axis=0))
print( np.cumprod(a2, axis=1))
```

## dot()/matmul(): 점곱/행렬곱 계산

```python
print(a2)
b2 = np.ones_like(a2)
print(b2)

print(np.dot(a2, b2))
print(np.matmul(a2, b2))
```

## tensordot(): 텐서곱 계산

```python
print(a2)
print(b2)
print('axes=default')
print(np.tensordot(a2, b2))
print()
print('axes=0')
print(np.tensordot(a2, b2, axes=0))
print()
print('axes=1')
print(np.tensordot(a2, b2, axes=1))
```

## cross(): 벡터곱

```python
x = [1, 2, 3]
y = [4, 5, 6]
print(np.cross(x, y))
```

## inner()/outer(): 내적/외적

```python
print(a2)
print(b2)
print(np.inner(a2, b2))
print(np.outer(a2, b2))
```

## mean(): 평균 계산

```python
print(a2)
print(np.mean(a2))
print(np.mean(a2, axis=0))
print(np.mean(a2, axis=1))
```

## std(): 표준 편차 계산

```python
print(a2)
print(np.std(a2))
print(np.std(a2, axis=0))
print(np.std(a2, axis=1))
```

## var(): 분산 계산

```python
print(a2)
print(np.var(a2))
print(np.var(a2, axis=0))
print(np.var(a2, axis=1))
```

## min(): 최소값

## max(): 최대값

```python
print(a2)
print(np.max(a2))
print(np.max(a2, axis=0))
print(np.max(a2, axis=1))
```

## argmin(): 최소값 인덱스

## argmax(): 최대값 인덱스

## median(): 중앙값

```python
print(a2)
print(np.median(a2))
print(np.median(a2, axis=0))
print(np.median(a2, axis=1))
```

## percentile(): 백분위 수



image.png

```python
a1 = np.array([0, 1, 2, 3])
print(a1)
print(np.percentile(a1, [0, 20, 40, 60, 80, 100], interpolation='linear'))
print(np.percentile(a1, [0, 20, 40, 60, 80, 100], interpolation='higher'))
print(np.percentile(a1, [0, 20, 40, 60, 80, 100], interpolation='lower'))
print(np.percentile(a1, [0, 20, 40, 60, 80, 100], interpolation='nearest'))
print(np.percentile(a1, [0, 20, 40, 60, 80, 100], interpolation='midpoint'))
```

## any()

```python
a2 = np.array([[False, False, True],
               [True, True, True],
               [False, True, True]])

print(a2)

print(np.any(a2))
print(np.any(a2, axis=0))
print(np.any(a2, axis=1))
```

## all()

```python
print(a2)

print(np.all(a2))
print(np.all(a2, axis=0))
print(np.all(a2, axis=1))
```
