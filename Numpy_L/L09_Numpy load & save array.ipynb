## 배열 입출력


```python
a2 = np.random.randint(1, 10, size=(5, 5))
print(a2)
np.save('a', a2)
```

```python
print(a2)
np.savetxt("a.csv", a2, delimiter=',') # csv 파일이니까 , 로 설정
```

```python
# COLAB
# !ls 

# Local
import os
os.listdir()
```

```python
f = open("a.csv", 'r')
for i in f:
    print(i)
f.close()
```

```python
csv = np.loadtxt('a.csv', delimiter=',')
csv
```

```python
print(b2)
np.savetxt("b.csv", b2, delimiter=',', fmt='%.2e', header='c1, c2, c3, c4, c5') # fmt 소숫점 자릿수,
```

```python
f = open("b.csv", 'r')
for i in f:
    print(i)
f.close()
```

```python
csv = np.loadtxt('b.csv', delimiter=',')
print(csv)
```
