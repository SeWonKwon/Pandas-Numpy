# Step 1. Import the necessary librariesÂ¶

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
%matplotlib inline
```

# Step 2. Create the DataFrame it should look like below.

```python
colnames = ['first_name', 'last_name', 'age', 'female', 'preTestScore', 'postTestScore']

data = {'first_name':['Jason', 'Molly','Tina','Jake','Amy'],
            'last_name':('Miller','Jacobson','Ali','Milner','Cooze'),
            'age':(42, 52, 36, 24, 73),
            'female':(0, 1, 1, 0, 1),
            'preTestScore':(4, 24, 31, 2, 3),
            'postTestScore':(25, 94, 57, 62, 70)}


df = pd.DataFrame(data, columns=colnames)
df
```

Step 3. Create a Scatterplot of preTestScore and postTestScore, with the size of each point determined by age
Hint: Don't forget to place the labels

```python
sns.scatterplot(x='preTestScore', y='postTestScore', data=df, size='age');
```

```python
plt.scatter(x=df.preTestScore, y=df.postTestScore,
           s=df.age);
```

Step 4. Create a Scatterplot of preTestScore and postTestScore.
This time the size should be 4.5 times the postTestScore and the color determined by sex

```python
sns.scatterplot(x='preTestScore', y='postTestScore',hue='female', data=df, size=df.postTestScore ,sizes=(100 ,300) );
```

```python
plt.scatter(x=df.preTestScore, y=df.postTestScore,
           s=df.postTestScore*4.5, c=df.female,cmap='viridis');
```
