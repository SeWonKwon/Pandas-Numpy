# Step 6. What was the quantity of the most expensive item ordered?   
 가장 비싼 아이템은 주문량은 어떻게 되나요?

```python
# item_price 를 float 형으로 바꾸기 위해서 $ 를 제거하기
dollarizer = lambda x : float(x[1:-1])
chipo['item_price_d'] = chipo.item_price.apply(dollarizer)
```

```python
chipo.loc[:,['item_price','item_price_d']]
```

```python
chipo.sort_values('item_price_d', ascending=False)
```

```python
list(chipo.sort_values('item_price_d', ascending=False).head(1).item_name)[0]
```

# Step 8. How many times people orderd more than one Canned Soda?   
소다 한캔 이상을 주문은 몇번 이루어 졌나요?

```python
list(chipo.choice_description.unique()) # choice 에는 Soda가 없음
```

```python
item_list = chipo.item_name.unique()
item_list
```

```python
for i in item_list:
    if ('soft' in i.lower()) & ('tacos' not in i.lower()):
        print(i)
    if ('soda' in i.lower()) & ('tacos' not in i.lower()):
        print(i)
```

```python
drink_list = []
for i in item_list:
    if ('soft' in i.lower()) & ('tacos' not in i.lower()):
        print(i)
        drink_list.append(i)
    if ('soda' in i.lower()) & ('tacos' not in i.lower()):
        print(i)
        drink_list.append(i)
```

```python
drink_list
```

```python
chipo.item_name.isin(drink_list)
```

```python
chipo.item_name.isin(drink_list).sum()
```
