# Series 선언

* index 와 value 값을 설정 해야 한다. 
    - index를 미 할당시 자동 RangeIndex로 설정된다.

```python
s = pd.Series([0, 0.25, 0.5, 0.75, 1.0],
              index=['a','b','c','d','e'])
s
```

```python
rs = pd.Series(np.random.randint(1,5,10))
rs
```

```python
rs.index
```

```python
rs_f = pd.Series(np.random.randint(1,5,10), dtype='float')
rs_f
```

# Series index

시리즈는 인덱스만 가지고 있다. 

```python
s = pd.Series([0, 0.25, 0.5, 0.75, 1.0],
              index=['a','b','c','d','e'])
s
```

```python
s['c']
```

* Series는 인덱스별로 직접 접근이 가능하다.
    * DataFrame 은 **컬럼별로만** 직접 접근이 가능하다.

```python
# 직접 접근
s[['c','d','e']]
```

```python
'b' in s.index
```

```python
s['a':'d']
```

## isin()

* 시리즈 객체끼리만 가능하기 때문에 양쪽다 씨리즈여야 한다. 

```python
s1 = pd.Series(np.random.randint(1,10,10))
s2 = pd.Series(np.random.randint(1,10,10))
```

```python
s1.isin(s2)
```

```python
s.isin([0.25, 0.75])
```

## 사칙연산

\*   /  +  -  **  % //

```python
s1 = pd.Series(np.random.randint(1,10,10))
s2 = pd.Series(np.random.randint(1,10,10))
```

```python
s1, s2
```

```python
s1*s2
```

```python
s1//s2, s1%s2, s1**s2
```
