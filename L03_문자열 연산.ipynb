## 문자열 연산

#### 문자열 연산자

```python
name_tuple = ['Suan Lee', 'Steven Jobs', 'Larry Page',
              'Elon Musk', None, 'Bill Gates', 
              'Mark Zuckerbuerg', 'Jeff Bezos']
names = pd.Series(name_tuple)
names
```

```python
names.str.lower()
```

```python
names.str.len()
```

```python
names.str.split()
```

#### 기타 연산자


```python
names.str[0:4]
```

```python
names.str.split().str.get(-1) # 성만 가져오기
```

```python
names.str.repeat(2)
```

```python
names.str.join('*')
```

#### 정규표현식


```python
names.str.match('([A-Za-z]+)')
```

```python
names.str.findall('([A-Za-z]+)')
```
